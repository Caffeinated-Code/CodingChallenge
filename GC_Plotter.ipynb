```python
from Bio.Seq import Seq
from sys import exit
from Bio.SeqUtils import GC_skew
import matplotlib.pyplot as plt
```

```python
inSeq = input("Please enter a nucleotide sequence: ")
```

```python
"""Chomp end spaces"""
inSeq = str.rstrip(inSeq) 
"""Chomp whitespaces in the middle"""
inSeq = inSeq.replace(' ', '') 
print(inSeq)
```

```python
def check(test_str):
    import re
    pattern = r'[^\.ATGCUatgcu]'
    if re.search(pattern, test_str, flags=re.IGNORECASE):
        print('Invalid nucleotide sequence : %r' % (test_str,)) 
        return(1)
    else:
        print ('Valid nucleotide sequence : %r' % (test_str,))
        return(0)
```

```python
if check(test_str=inSeq) == 1:
    exit()
```

```python
inSeq = inSeq.upper()
g = inSeq.count("G")
c = inSeq.count("C")
gc = ((g+c)/len(inSeq))*100
print("GC content of input sequence: ",gc)
```

```python
gcContent = GC_skew(inSeq, window=3)
```

```python
plt.plot(gcContent, color='r', linewidth=2.0)
plt.ylabel('GC%')
plt.xlabel('Window')
plt.rc('axes', labelsize=30)  
plt.show()
```
